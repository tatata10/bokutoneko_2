```python
import requests
from bs4 import BeautifulSoup
import re
import json

# 取得したデータを保存する関数
def save_to_json(data, filename="machine_data.json"):
    with open(filename, "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=2)

def get_machine_dict(url):
    """
    指定された Anytime Fitness の施設ページから
    マシンのカテゴリごとのマシン名リストを辞書形式で取得する
    """
    res = requests.get(url)
    res.encoding = res.apparent_encoding  # 日本語対応
    soup = BeautifulSoup(res.text, "html.parser")

    machine_dict = {}
    for section in soup.find_all("div", class_=lambda x: x and "machine-item" in x):
        area_name = section.h2.get_text(strip=True)
        machine_list = []
        name_dict = {}
        for machine in section.find_all("li"):
            machine_name = machine.get_text(strip=True)
            name, count = parse_machine(machine_name)
            name_dict[name] = count
            machine_list.append(dict)
        machine_dict[area_name] = name_dict

    return machine_dict

def get_shop_name(url):
    res = requests.get(url)
    res.encoding = res.apparent_encoding
    soup = BeautifulSoup(res.text, "html.parser")

    # shop-name クラスを持つ span タグを取得
    shop_name_container = soup.find("span", class_="shop-name")
    if shop_name_container:
        inner_spans = shop_name_container.find_all("span")
        if len(inner_spans) >= 2:
            return inner_spans[1].text.strip()
    return None

def parse_machine(text):
    """
    例:
    'パワーラック×2' → ('パワーラック', 2)
    'パワーラック3台' → ('パワーラック', 3)
    'ダンベル（1kg〜40kg）' → ('ダンベル（1kg〜40kg）', 1)
    """
    # パターン①: ×2, x2, X2, ＊2, *2
    pattern1 = re.compile(r"(.+?)[×xX＊*](\d+)")
    match1 = pattern1.search(text)
    if match1:
        name = match1.group(1).strip()
        count = int(match1.group(2))
        return name, count

    # パターン②: 3台, 1台, 10台など
    pattern2 = re.compile(r"(.+?)(\d+)台")
    match2 = pattern2.search(text)
    if match2:
        name = match2.group(1).strip()
        count = int(match2.group(2))
        return name, count

    # 台数が明記されていない場合は1台とみなす
    return text.strip(), 1

def get_facility_urls_from_prefecture(prefecture_url):
    """
    指定された都道府県ページから、各店舗の facility ページの URL を取得する
    例: https://www.anytimefitness.co.jp/kanto/tokyo/
    """
    res = requests.get(prefecture_url)
    res.encoding = res.apparent_encoding
    soup = BeautifulSoup(res.text, "html.parser")

    store_url_list = []
    store_name_list = []
    for section in soup.find_all("div", class_=lambda x: x and "areasort" in x):
        for store_list in section.find_all("li"):
            a_tag = store_list.find("a", href=True)
            store_name = store_list.find("div", class_=lambda x: x and "name" in x)
            if a_tag:
                store_path = a_tag["href"]  # 例: shimomaruko/
                facility_url = "https://www.anytimefitness.co.jp" + store_path + "facility/"
                store_url_list.append(facility_url)
                store_name_list.append(store_name)

    return store_url_list,store_name_list

if __name__ == "__main__":
    tokyo_url = "https://www.anytimefitness.co.jp/kanto/tokyo/"
    facility_urls,store_names = get_facility_urls_from_prefecture(tokyo_url)
    total = len(facility_urls)
    all_data = []

    for i, (url, name) in enumerate(zip(facility_urls, store_names), start=1):
        print(f"[{i}/{total}] {round(i/total*100, 1)}% 完了: {url}")
        try:
            machine_data = get_machine_dict(url)
            store_name = get_shop_name(url)
            all_data.append({
                "store_name": store_name,
                "machines": machine_data
            })
        except Exception as e:
            print(f"⚠️ エラー: {e}")

    save_to_json(all_data, "tokyo_machines.json")
    print("✅ JSONファイルに保存しました: tokyo_machines.json")
```
